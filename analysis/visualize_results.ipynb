## Likert scale questions

```python
# Get question pairs (excluding the open-ended questions and demographic data)
trivial_cols = [col for col in df.columns if "(Système trivial)" in col and "[" in col]
our_cols = [col for col in df.columns if "(Notre système)" in col and "[" in col]

print(trivial_cols)
print(our_cols)
```

```python
import re

pattern = r"\[(.*)\]"


def get_base_question(full_col_name):
    return re.search(pattern, full_col_name).group(1)


questions = []
for col in trivial_cols:
    question = get_base_question(col)
    questions.append(question)
    print(question)

print(len(questions))
```

```python
import scipy.stats as stats
import numpy as np


def compute_metrics(df):
    metrics = {}

    for col in df.columns:
        data = df[col].dropna()

        metrics[col] = {
            "mean": np.mean(data),
            "median": np.median(data),
            "std": np.std(data),
            "variance": np.var(data),
        }

    # Mann-Whitney U test
    statistic, pvalue = stats.mannwhitneyu(
        df["Système trivial"].dropna(),
        df["Notre système"].dropna(),
        alternative="two-sided",
    )

    # d = (mean(x1) – mean(x2)) / sqrt((s1^2 + s2^2) / 2)
    # Cohen's d
    cohens_d = (
        metrics["Notre système"]["mean"] - metrics["Système trivial"]["mean"]
    ) / np.sqrt(
        (
            metrics["Notre système"]["variance"] * metrics["Notre système"]["variance"]
            + metrics["Système trivial"]["variance"]
            * metrics["Système trivial"]["variance"]
        )
        / 2
    )

    metrics["comparison"] = {
        "mannwhitney_stat": statistic,
        "mannwhitney_p": pvalue,
        "cohens_d": cohens_d,
    }

    return metrics


def display_metrics(metrics, question):
    print(f"\n=== {question} ===")
    print(
        f"Trivial system: mean={metrics['Système trivial']['mean']:.2f}, std={metrics['Système trivial']['std']:.2f}, median={metrics['Système trivial']['median']:.2f}"
    )
    print(
        f"Our system: mean={metrics['Notre système']['mean']:.2f}, std={metrics['Notre système']['std']:.2f}, median={metrics['Notre système']['median']:.2f}"
    )

    # Interpret statistical significance
    p_val = metrics["comparison"]["mannwhitney_p"]
    cohens = metrics["comparison"]["cohens_d"]

    if p_val < 0.05:
        print(f"Significant difference (p={p_val:.3f})")
        print(
            f"Effect size: {cohens:.2f} ({'large' if abs(cohens) > 0.8 else 'medium' if abs(cohens) > 0.5 else 'small'})"
        )
    else:
        print("No significant difference")
        print(
            f"Effect size: {cohens:.2f} ({'large' if abs(cohens) > 0.8 else 'medium' if abs(cohens) > 0.5 else 'small'})"
        )
```

```python
import textwrap

# Calculate number of rows needed (2 plots per row)
n_cols = 5
n_rows = 5

# Create a figure with a grid of subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 4 * n_rows))
# Flatten axes array for easier iteration
axes = axes.flatten()

all_metrics = {}

# Draw boxplots for each question
for idx, question in enumerate(questions):
    trivial_col = [col for col in trivial_cols if question in col][0]
    our_col = [col for col in our_cols if question in col][0]

    # Create a DataFrame with custom labels
    plot_data = pd.DataFrame(
        {"Système trivial": df[trivial_col], "Notre système": df[our_col]}
    )

    metrics = compute_metrics(plot_data)
    display_metrics(metrics, question)
    all_metrics[question] = metrics

    # Plot the boxplot in the corresponding subplot
    sns.boxplot(data=plot_data, ax=axes[idx], native_scale=True)
    axes[idx].set_title("\n".join(textwrap.wrap(question, width=33)))

    # Set integer ticks from 1 to 5
    axes[idx].set_yticks([1, 2, 3, 4, 5])
    # Set y-axis limits to compress the plot
    axes[idx].set_ylim(0.5, 5.5)


# fig.delaxes(axes[-3])
# fig.delaxes(axes[-2])
# fig.delaxes(axes[-1])

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

# Create latex table

```python
import pyperclip

data = []
for metric, values in all_metrics.items():
    # Determine which mean should be bolded
    trivial_mean = values["Système trivial"]["median"]
    our_mean = values["Notre système"]["median"]

    bold_one = False
    if abs(trivial_mean - our_mean) > 0.01:
        bold_one = True
        # Special case for "Je me suis senti perdu.e"
        if "Je me suis senti perdu" in metric:
            bold_trivial = trivial_mean < our_mean
        else:
            bold_trivial = trivial_mean > our_mean
    else:
        bold_one = False

    if bold_one:
        # Add bold tags to the larger value (or smaller for "perdu.e")
        trivial_mean_str = (
            f"\\textbf{{{trivial_mean:.2f}}}" if bold_trivial else f"{trivial_mean:.2f}"
        )
        our_mean_str = (
            f"\\textbf{{{our_mean:.2f}}}" if not bold_trivial else f"{our_mean:.2f}"
        )
    else:
        trivial_mean_str = f"{trivial_mean:.2f}"
        our_mean_str = f"{our_mean:.2f}"

    # Rename the columns and organize metrics in a table
    data.append(
        {
            "Question": metric,
            "\\bar{x} (T)": values["Système trivial"]["mean"],
            "M (T)": trivial_mean_str,
            "\\sigma (T)": values["Système trivial"]["std"],
            "\\bar{x} (O)": values["Notre système"]["mean"],
            "M (O)": our_mean_str,
            "\\sigma (O)": values["Notre système"]["std"],
            "\\textit{p}": values["comparison"]["mannwhitney_p"],
            "\\textit{d}": values["comparison"]["cohens_d"],
        }
    )

table_df = pd.DataFrame(data)

latex_table = table_df.to_latex(
    index=False,
    float_format="%.2f",  # format floats to 2 decimal places
    column_format="p{6cm}|r{1.5cm}r{1.5cm}r{1.5cm}|r{1.5cm}r{1.5cm}r{1.5cm}|r{1.5cm}r{1.5cm}",
    escape=False,  # allows LaTeX commands to be passed through
)


# Add multicolumn headers
latex_table = latex_table.replace("\\begin{tabular}", "\\begin{tabular}[t]")

# Insert the multicolumn headers after the first line
header_line = (
    "\\hline\n"
    "& \\multicolumn{3}{c|}{Système trivial} & \\multicolumn{3}{c|}{Notre système} & \\multicolumn{2}{c}{Tests} \\\\\n"
)

# Update column headers to remove (T) and (O)
latex_table = latex_table.replace(
    "Question & \\bar{x} (T) & M (T) & \\sigma (T) & \\bar{x} (O) & M (O) & \\sigma (O)",
    header_line + "Question & \\bar{x} & M & \\sigma & \\bar{x} & M & \\sigma",
)

# Add extra hline after the column headers
latex_table = latex_table.replace("\\\\\\hline\n", "\\\\\\hline\\hline\n")


print(latex_table)

# Copy to clipboard for convenience
pyperclip.copy(latex_table)
```
